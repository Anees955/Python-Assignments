```python
#todo List

list_todo =[]
userInput = input("\tWELCOME TO TODO LIST \n1 to Add elements \n2 to remove specific \n3 to remove all \nq or Q to Quit\n")

while True :
    
    if userInput == "q" or userInput == "Q":
        print("-"*10,"TODO LIST","-"*10)
        for x in list_todo:
            print(list_todo.index(x)+1,x)
        print("-"*13,"END","-"*13)
        break
        
    elif userInput == "1" :
        addInput = input("\nAdd items As many as you want\n")
        list_todo.append(addInput)
        print("-"*10,"TODO LIST","-"*10)
        for x in list_todo:
            print(list_todo.index(x)+1,x)
        print("-"*13,"END","-"*13)
        userInput = input("\nItem Added!\n\n1 to Add more  \n2 To delete specific value from list \n3 to delete all \nq or Q to quit\n")

    elif userInput == "2" :
        
        print("-"*10,"TODO LIST","-"*10)
        for x in list_todo:
            print(list_todo.index(x)+1,x)
        print("-"*13,"END","-"*13)
        delInput = int(input("Type element number to remove eg. 1 \n"))
        delInput = delInput-1
        del list_todo[delInput]
        #list_todo.remove(delInput)
        print("-"*10,"TODO LIST","-"*10)
        for x in list_todo:
            print(list_todo.index(x)+1,x)
        print("-"*13,"END","-"*13)
        userInput = input("\nItem Removed!\n\n1 to Add more  \n2 To delete specific value from list \n3 to delete all \nq or Q to quit\n")
    
    elif userInput == "3" :
        list_todo.clear()
        print("-"*10,"TODO LIST","-"*10)
        for x in list_todo:
            print(list_todo.index(x)+1,x)
        print("-"*13,"END","-"*13)
        userInput = input("\nList has been deleted!\n\n1 to Add items  \n2 To delete specific value from list \n3 to delete all \nq or Q to quit\n")
        
        
```
