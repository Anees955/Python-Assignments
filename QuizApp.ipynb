```python
#GK Quiz App Using List of Dict:

quiz = [
    {
        "Q_No." : "1",
        "Question" : "What is the capital city of pakistan?",
        "Options" : ["Karachi","Hyderabad","Islamabad","Lahore"],
         "Answer" : "Islamabad"
    },
    {
        "Q_No." : "2",
        "Question" : "What is the area of capital city of pakistan?",
        "Options" : ["930 km","1033 km","960 km","1400 km"],
         "Answer" : "960 km"
    },
    {
        "Q_No." : "3",
        "Question" : "Which one is the biggest city of pakistan?",
        "Options" : ["Karachi","Hyderabad","Islamabad","Lahore"],
         "Answer" : "Karachi"
    },
    {
        "Q_No." : "4",
        "Question" : "Which one is coldest city of pakistan?",
        "Options" : ["Marri","Skardu","Islamabad","Lahore"],
         "Answer" : "Skardu"
    },
    {
        "Q_No." : "5",
        "Question" : "How many Disctricts are in Sindh, pakistan?",
        "Options" : ["10","7","5","9"],
         "Answer" : "7"
    }
]

count = 0

for que in quiz:
    data = "\n\n"+que["Q_No."] +": "+ que["Question"] +"\n"+ que["Options"][0] +"\n"+ que["Options"][1] +"\n"+ que["Options"][2] +"\n"+ que["Options"][3] +"\n\nMake a Choice Just type A, B, C or D"+"\n\n"
    
    ans = input(data)
    if ans == "a" or ans == "A" :
        choice=que["Options"][0]
        ans = que["Answer"]
        if choice==ans:
            count += 1   
    elif ans == "b" or ans == "B" :
        choice=que["Options"][1]
        ans = que["Answer"]
        if choice==ans:
            count += 1
    elif ans == "c" or ans == "C":
        choice=que["Options"][2]
        ans = que["Answer"]
        if choice==ans:
            count += 1
    elif ans == "d" or ans == "D" :
        choice=que["Options"][3]
        ans = que["Answer"]
        if choice==ans:
            count += 1
            
                                                    #Sir i have make this more easy for user now he just have to type A B C or D
                                                        # intead of typing whole Option name typing whole options name in input can
                                                        # cause a problem because this is case sensitive
    #if ans == que["Answer"]:
        #count += 1
        
obt_Marks=count
quiz_Total_Score = 5
perc = (obt_Marks/quiz_Total_Score)*100
print("your Score is: ",count," Out of ",quiz_Total_Score, "\nPercentage is: ",perc,"%")
```
