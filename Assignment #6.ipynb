```python
"""Question 5:
Write a code in python in which create a class named it Car which
have 5 attributes such like (model, color and name etc.) and 3
methods. And create 5 object instance from that class. """
```

```python
#Question #5
#Solution

class Car():
    def __init__(self,make,model,speed,color,seats):
        
        self.make = make
        self.model = model
        self.speed = speed
        self.color = color
        self.seats = seats
        
        
    def start(self):
        print(f"your Car {self.make} is Started! \nEnjoy Your Ride!\n")
        
    def stop(self):
        print(f"your Car {self.make} is Stoped!\n")
    
    def display(self):
        print(f"This is a {self.make} {self.model}\nhaving Speed {self.speed} \nColor {self.color} \nSeats {self.seats}\n")
    
obj1 = Car("Honda","2019","370KMPH","Blue","4")  
obj1.start()
obj1.display()
obj1.stop()


obj2 = Car("Toyota","2019","350KMPH","White","4")  
obj2.display()

obj3 = Car("Ferrari","2019","400KMPH","Red","2")  
obj3.display()
```
